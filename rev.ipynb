```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans,DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
X,y_true = make_blobs(n_samples=500,centers=3,cluster_std=0.60,random_state=42)
df = pd.DataFrame(X,columns=["Feature1","Feature2"])
df.head()
```

```python
scaler = StandardScaler()
l = ["Feature1","Feature2"]
X_scaler = df
X_scaler[l] = scaler.fit_transform(X_scaler[l])
X_scaler.head() 
```

```python
inertia = []
K_range = range(1,11)
for k in K_range:
    k_means = KMeans(n_clusters=k,random_state=42)
    k_means.fit(X_scaler)
    inertia.append(k_means.inertia_)
inertia
```

```python
plt.plot(K_range,inertia,marker="o")
plt.show()
```

```python
k_means = KMeans(n_clusters=3,random_state=42)
pred = k_means.fit_predict(X_scaler)
pred
```

```python
df["cluster"] = pred
df.head()
```

```python
sns.scatterplot(df,x="Feature1",y="Feature2",hue="cluster",palette="viridis")
plt.show()
```

```python
from sklearn.datasets import make_moons
X,y_true = make_moons(n_samples=500,noise=0.05,random_state=42)
df = pd.DataFrame(X,columns=["Feature_1","Feature_2"])
df.head()
```

```python
l = ["Feature_1","Feature_2"]
X_scaler = df
X_scaler[l] = scaler.fit_transform(df[l])
X_scaler
```

```python
inertia = []
for k in K_range:
    k_means = KMeans(n_clusters=k,random_state=42)
    k_means.fit(X_scaler)
    inertia.append(k_means.inertia_)
inertia
```

```python
plt.plot(K_range,inertia,marker="o")
plt.show()
```

```python
k_means = KMeans(n_clusters=2,random_state=42)
pred = k_means.fit_predict(X_scaler)
pred
```

```python
df["cluster"] = pred
df.head()
```

```python
sns.scatterplot(df,x="Feature_1",y="Feature_2",hue="cluster",palette="viridis")
plt.show()
```

```python
# df.head()
df = df.drop("cluster",axis=1)
df.head()
```

```python
dbscn = DBSCAN(eps=0.3,min_samples=5)
pred = dbscn.fit_predict(X_scaler[["Feature_1","Feature_2"]])
pred 
```

```python
df["cluster"] = pred
df.head()
```

```python
sns.scatterplot(df,x="Feature_1",y="Feature_2",hue="cluster",palette="viridis")
plt.show()
```

```python
X,y = make_blobs(n_samples=500,n_features=5,centers=3,cluster_std=1.5,random_state=42)
y
```

```python
X
```

```python
X_scaler = scaler.fit_transform(X)
X_scaler
```

```python
pca = PCA(n_components=2)
pca_trans = pca.fit_transform(X_scaler)
pca_trans
```

```python
df = pd.DataFrame(pca_trans,columns=["PCA1","PCA2"])
df["label"] = y
df.head()
```

```python
sns.scatterplot(df,x="PCA1",y="PCA2",hue="label",palette="Set2")
plt.show()
```
