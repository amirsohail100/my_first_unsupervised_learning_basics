```python
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame({
    "Customer":["riya","aman","amir","harry","sneha","faizan"],
    "Age":[20,30,40,22,38,25],
    "Spending":[100,200,300,110,290,130]
})
X = df[["Age"]]
model = KMeans(n_clusters=2,random_state=42,n_init=20)
df["group"] = model.fit_predict(X)

plt.figure(figsize=(6,5))
for group in df["group"].unique():
    group_Data = df[df["group"] == group]
    plt.scatter(group_Data["Age"],group_Data["Spending"],label = f"Group{group}")
    sns.scatterplot(x=group_Data["Age"],y=group_Data["Spending"],)
    print(group_Data)
plt.xlabel("Age")
plt.xlabel("Spending")
plt.legend()
plt.grid()
plt.show()
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.DataFrame({
    "Age":[25,30,35,40,45,50],
    "Income":[30000,40000,50000,60000,70000,80000],
    "Spending":[70,60,50,40,30,20],
    "Savings":[1000,5000,8000,10000,15000,20000]
})
Scaler = StandardScaler()
Standard_Scaler = Scaler.fit_transform(df)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(Standard_Scaler)
pca_df = pd.DataFrame(pca_result,columns=["PCA1","PCA2"])
explained_Var = pca.explained_variance_ratio_
print(np.round(explained_Var * 100,2))
plt.figure(figsize=(8,6))
plt.scatter(pca_df["PCA1"],pca_df["PCA2"])
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid()
plt.show()
pca_df
```
