```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X,y = make_blobs(n_samples=500,n_features=5,centers=3,cluster_std=1.5,random_state=42)
X
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
X_scaler
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X_scaler)
x_pca
```

```python
df_pca = pd.DataFrame(x_pca,columns=["PCA_1","PCA_2"])
df_pca["label"] = y
df_pca
```

```python
sns.scatterplot(x=df_pca["PCA_1"],y=df_pca["PCA_2"],hue=df_pca["label"],palette="Set2")
plt.show()
```
